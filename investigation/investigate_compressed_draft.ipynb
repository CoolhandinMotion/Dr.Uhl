```python
import re
import numpy as np
import pandas as pd
import pandas
from scipy import stats as st
import imagecodecs as imcs
import cv2
import glob
import matplotlib.pyplot as plt
from pathlib import Path
import os
import statistics as st
from numpy.typing import NDArray
```

```python
ROOT_DIR = Path(os.getcwd())
COMPRESSED_DATASET_DIR = ROOT_DIR / 'compressed_draft'
compressed_img_paths = glob.glob(f'{COMPRESSED_DATASET_DIR}/*.jpeg')
```

```python
compressed_img_sizes = [os.path.getsize(img_path) for img_path in compressed_img_paths]
compressed_img_sizes = np.asarray(compressed_img_sizes)//1000
print(f"number of compressed images are {len(compressed_img_sizes)}")
print(f"minimum image size is: {min(compressed_img_sizes)} KB")
print(f"maximum image size is: {max(compressed_img_sizes)} KB")
```

```python
fig, ax = plt.subplots()
ax.boxplot(compressed_img_sizes)
ax.set_title("Box plot of Compressed image Sizes")
ax.set_ylabel("Kilobyte")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(compressed_img_sizes,bins=70)
ax.set_title("Histogram of Compressed image sizes")
ax.set_xlabel('Kilobyte')
ax.set_ylabel('numbers in bin')
plt.show()
```
